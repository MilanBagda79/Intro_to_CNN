# <b>Introduction to Convolutional Neural Networks</b>

## <b>Definition of Convolutional Neural Networks (CNN)</b>

Convolutional Neural Networks (CNNs) are a category of Neural Networks that have proven very effective in areas such as image recognition and classification. Pioneered by Yann LeCun in the 1980s, CNNs are used in many of the AI applications that we see in our day-to-day lives.

A Convolutional Neural Network is designed to automatically and adaptively learn spatial hierarchies of features from the training data. It is made up of neurons that have learnable weights and biases, just like a normal Neural Network. However, the architecture of a CNN is designed to take advantage of the 2D structure of an input image (or other 2D input such as a speech signal).

## <b>Why CNN for Image Data?</b>
CNNs are particularly suited for analyzing images for several reasons:

* <b>Local Invariance:</b> CNNs automatically learn and generalize over spatial translations in the data, which means they can recognize a pattern or object in an image, regardless of its location in the image.

* <b>Compositionality:</b> Each filter composes a local patch of lower-level features into higher-level representation. This allows the network to learn increasingly complex visual concepts, since real-world images are made up of compositional hierarchies where high-level features are made of an arrangement of lower-level ones.

* <b>Reduced Complexity:</b> Compared to fully connected networks, CNNs have significantly fewer parameters, making them less prone to overfitting and more computationally efficient.

## <b>Applications of CNN</b>

CNNs are mainly used in image processing and computer vision tasks, but their application areas are expanding to other types of data. Some of the notable applications are:

* <b>Image and Video Recognition:</b> CNNs can be used to categorize and identify objects and even people in images and videos.

* <b>Image Segmentation:</b> CNNs are effective in tasks like semantic segmentation, where the goal is to classify each pixel in the image as belonging to a particular class.

* <b>Medical Image Analysis:</b> CNNs are used in identifying anomalies and diseases in medical images such as MRI scans or X-ray images.

* <b>Self-driving Cars:</b> Autonomous vehicles use CNNs to detect objects, pedestrians, and other vehicles to navigate safely.

* <b>Facial Recognition:</b> CNNs are used in facial recognition systems for identity verification, emotion detection, and even to track persons of interest in surveillance systems.

## <b>Key Components of CNNs</b>

CNNs are made up of several types of layers, each playing a distinct role in the learning process. The three main types of layers in a CNN are:

### <b>Convolutional Layer</b>

The convolutional layer is the core building block of a CNN. The layer's parameters consist of a set of learnable kernels or filters, which have a small receptive field, but extend through the full depth of the input volume. During the forward pass, each filter is convolved across the width and height of the input volume, computing the dot product between the entries of the filter and the input, producing a 2-dimensional activation map. As a result, the network learns filters that activate when they see some type of visual feature such as an edge or a blob of color in the input.

### <b>Pooling Layer</b>

The pooling (or subsampling) layer is usually inserted between successive convolutional layers in a CNN architecture. Its function is to progressively reduce the spatial size of the input volume, to reduce the amount of parameters and computation in the network, and hence to also control overfitting. The pooling layer operates independently on each depth slice of the input and resizes it spatially using the max or average operation.

### <b>Fully Connected Layer</b>

The Fully Connected layer is a traditional Multi Layer Perceptron that uses a softmax activation function in the output layer (or in other words, the weights from the last feature map are fully connected to the output). The purpose of the Fully Connected layer is to use these features for classifying the input image into various classes based on the training dataset. Apart from classification, adding a fully-connected layer is also a (usually) cheap way of learning non-linear combinations of the high-level features as represented by the output of the convolutional layer.

## <b>Architecture of a Basic CNN</b>

A basic Convolutional Neural Network consists of a stack of Convolutional Layers, followed by Pooling Layers. The output of these layers is flattened and passed through one or more Fully Connected Layers. The final Fully Connected layer, often accompanied by a softmax activation function, provides the output of the network.

### <b>Convolution + ReLU Layer</b>

In the convolution layer, various filters/kernels slide across the image and produce a feature map. This process is known as convolution. The ReLU (Rectified Linear Unit) Layer applies the activation function element-wise. It is used after every Convolutional Layer in a CNN. An important aspect of ReLU is that it introduces non-linearity into the model, a necessary component for learning from complex datasets.

### <b>Pooling Layer</b>

As explained before, the role of a pooling layer is to reduce the spatial dimensions (width & height) of the input volume for the next convolutional layer. It does not affect the depth dimension of the volume.

### <b>Flattening</b>

Flattening is the process of converting all the resultant 2-dimensional arrays from pooled feature maps into a single long continuous linear vector (which serves as the input for the Fully Connected layer).

### <b>Fully Connected Layer</b>

A fully connected layer takes the inputs, as derived from the previous layer, and calculates the class probabilities. The output of this layer is equivalent to the final predicted class scores.

### <b>Output Layer</b>

The output layer of a CNN represents the network's predicted output. It typically consists of softmax, sigmoid or other types of activation function, which squashes the output probabilities between the required range (usually 0 to 1). Each node in the output layer represents a particular class label, and the node with the highest activation represents the predicted class label.


## **Training a CNN**

Training a Convolutional Neural Network involves several steps, which we will go through below:

### **Forward Propagation**

Forward propagation in a CNN is the process where the input image is passed through the layers of the network. Each layer performs a specific operation on the input data (convolution, pooling, etc.), and passes its output to the next layer. The convolution operation performed in the convolution layer can be represented mathematically as:

`I * K + b`

where * represents the convolution operation, `I` is the input to the layer, `K` is the kernel or filter, and `b` is the bias term.

The final output is a probability that tells us what the input image represents.


### **Cost Function**

The cost function (or loss function) of a CNN quantifies the difference between the predicted output and the actual output of the network. It's a measure of the error for each prediction. The cross-entropy loss for binary classification problems can be represented mathematically as:

`L = -y * log(y_hat) - (1 - y) * log(1 - y_hat)`

where y is the true label and y_hat is the predicted label. For multiclass classification, it is extended to the sum over all classes:

`L = -sum(y * log(y_hat))`


### **Backpropagation**

Backpropagation in a CNN is used to calculate the gradient of the loss function with respect to the weights in the network. The calculated gradient is then used in the weight update stage to adjust the weights and bias values in order to minimize the output error. The gradients are calculated by applying the chain rule from calculus, and are propagated backward through the network (hence the name "backpropagation"). This involves computing the derivative of the loss function with respect to the weights and biases. For example, if `L` is the loss function and `w` represents the weights, then the partial derivative of `L` with respect to `w` is represented as `∂L/∂w`.

### **Weight Update**

In the weight update stage, the weights and bias of the network are updated in order to minimize the output error. The update is performed using an optimization algorithm, most commonly stochastic gradient descent (SGD) or a variant of it. The weights are updated according to the rule:

`w = w - α ∂L/∂w`

where `α` is the learning rate, a hyperparameter that determines the step size during weight updates. This process is repeated for several epochs (iterations over the entire dataset) until the model's performance on a separate validation set stops improving.

### **Implementing Convolutional Neural Network for Image Recognition**

### **Import necessary libraries**

```python
import numpy as np
import pandas as pd
np.random.seed(42) 
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import SVG
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten,GlobalAveragePooling2D, Input, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D
from keras.layers import Convolution2D, MaxPooling2D, AveragePooling2D, GlobalMaxPooling2D, GlobalAveragePooling2D
from keras.utils import to_categorical
from keras.datasets import mnist
```

### **Load Mnist dataset**

```python
#  Load pre-shuffled MNIST data into train and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

### **Prepare the data for training**

```python
# IF YOU ARE USING YOUR OWN MACHINE, THEN YOU CAN TRY A LARGER DATASET BY COMMENTING THE BELOW CODE:
X_train = X_train[0:1000,]
y_train = y_train[0:1000,]
X_test = X_test[0:200,]
y_test = y_test[0:200,]
```

### **Preprocessing Input Data**

```python
# Preprocess input data
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0],28, 28, 1)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

```python
print('X_train shape:', X_train.shape)
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

### **Preprocessing Class Labels**

```python
#  Preprocess class labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

### **Displaying a Subset of the Training Images**

```python
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(X_train[i].reshape(28, 28), cmap='gray')
```

### **Defining the Model Architecture**

```python
# Define model architecture
model = Sequential()

# Add the first convolutional layer with 64 filters, a 3x3 kernel, and ReLU activation
model.add(Convolution2D(64, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)))

# Add another convolutional layer with 64 filters and a 3x3 kernel, followed by ReLU activation
model.add(Convolution2D(64, kernel_size=(3,3), activation='relu'))

# Add a max pooling layer with a 2x2 pool size to downsample the feature maps
model.add(MaxPooling2D(pool_size=(2,2)))

# Add more convolutional layers with 64 filters and 3x3 kernels, followed by ReLU activation
model.add(Convolution2D(64, kernel_size=(3,3), activation='relu'))
model.add(Convolution2D(64, kernel_size=(3,3), activation='relu'))

# Add another max pooling layer with a 2x2 pool size
model.add(MaxPooling2D(pool_size=(2,2)))

# Add a convolutional layer with 128 filters and a 3x3 kernel, followed by ReLU activation
model.add(Convolution2D(128, kernel_size=(3,3), activation='relu'))

# Apply global average pooling to reduce the spatial dimensions of the feature maps
model.add(GlobalAveragePooling2D())

# Apply dropout regularization with a rate of 0.25 to prevent overfitting
model.add(Dropout(0.25))

# Add a fully connected layer with 32 units and ReLU activation
model.add(Dense(32, activation='relu'))

# Apply dropout regularization with a rate of 0.5
model.add(Dropout(0.5))

# Add the final fully connected layer with 10 units (corresponding to the number of classes) and softmax activation
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

### **Compiling the Model**

```python
# Compile model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['categorical_accuracy'])
```

### **Fitting the Model on Training Data**

```python
# Fit model on training data
history= model.fit(X_train, y_train, 
          batch_size=32, epochs=10, verbose=1, validation_data=(X_test,y_test))
```

### **Evaluating the Model on Test Data**

```python
# Evaluate model on test data
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
print("Test Error: %.2f%%" % (100-score[1]*100))
```

### **Displaying the Predicted Labels on Test Images**

```python
# Use the trained model to predict the labels for the test dataset
res = np.argmax(model.predict(X_test), axis=-1)

# Create a figure to display the predicted images
plt.figure(figsize=(10, 10))

# Loop over the first 9 images in the test dataset
for i in range(9):
    # Create a subplot for each image
    plt.subplot(3, 3, i+1)
    
    # Display the image by reshaping it to 28x28 and using a grayscale colormap
    plt.imshow(X_test[i].reshape(28,28), cmap='gray')
    
    # Remove x-axis ticks
    plt.gca().get_xaxis().set_ticks([])
    
    # Remove y-axis ticks
    plt.gca().get_yaxis().set_ticks([])
    
    # Add a label showing the predicted class for the image
    plt.ylabel("prediction = %d" % res[i], fontsize=18)
```

### **Plotting the Training History**

```python
# list all data in history
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['categorical_accuracy'])
plt.plot(history.history['val_categorical_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
